```python
from geopy.geocoders import Nominatim

def get_country_from_city(latitude, longitude):
    # Initialiser l'API Nominatim
    geolocator = Nominatim(user_agent="geoapiExercises")
    
    # Obtenir la localisation à partir de la latitude et de la longitude
    location = geolocator.reverse(f"{latitude},{longitude}")
    
    # Extraire les informations de l'adresse
    address = location.raw['address']
    
    # Récupérer le nom du pays
    country = address.get('country', '')
    
    return country

# Exemple d'utilisation
latitude = "48.8566"
longitude = "2.3522"
country = get_country_from_city(latitude, longitude)
print("Le pays de la ville est :", country)
```
